Seaborn 함수 학습

```python
#sns.set(): seaborn의 전반적인 시각화 스타일을 설정하는 함수

import seaborn as sns
import matplotlib.pyplot as plt

#기본 스타일 설정
sns.set(style="whitegrid");

#간단한 산점도 그리기
sns.scatterplot(x=[1,2,3,4],y=[10,20,30,40]);
plt.show();
```

```python
#sns.scatterplot(): 산점도를 그리는 함수

#데이터 생성
data=sns.load_dataset("iris");


#산점도 그리기
sns.scatterplot(x="sepal_length",y="sepal_width",data=data,hue="species");
plt.show();
```

```python
#sns.lineplot(): 선 그래프를 그리는 함수

#데이터 생성
data=sns.load_dataset("tips");

#선 그래프 그리기
sns.lineplot(x="total_bill",y="tip",data=data);
plt.show();
```

```python
sns.barplot(x="day", y="total_bill", data=data)
plt.show()
```

```python
sns.histplot(data["total_bill"], kde=True)  # kde=True로 커널 밀도 추정 추가
plt.show()
```

```python
sns.boxplot(x=data["total_bill"])
plt.show()
```

```python
sns.violinplot(x=data["total_bill"])
plt.show()
```

```python
# 데이터 생성
data = sns.load_dataset("flights")

# 피벗 테이블 생성
pivot_data = data.pivot_table(index="month", columns="year", values="passengers")

# 히트맵 그리기
sns.heatmap(pivot_data, annot=True, cmap="coolwarm")
plt.show()
```

```python
# 데이터 생성
data = sns.load_dataset("iris")

# `iris` 데이터셋에 대한 pairplot 그리기
sns.pairplot(data, hue="species")
plt.show()
```

```python
# 데이터 생성
data = sns.load_dataset("iris")

# 상관 행렬 계산
corr_matrix = data.corr(numeric_only=True) 

# 히트맵 그리기
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm")
plt.show()
```
