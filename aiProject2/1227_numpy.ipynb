Numpy함수

```python
import numpy as np
arr=np.array([1,2,3]);
print(arr);

arr=np.arange(0,10,2);      #arange(시작값, 종료값, 증가) 
print(arr);

arr=np.random.rand(3,3)     #random: 난수 생성, rand: 0~1 사이의 난수 >3행 3열로 생성
print(arr);

arr=arr.reshape(1,9);       #배열의 shape 변경 >1행 9열로 변경
print(arr);

print(np.sum([1,2,3]));     #배열 요소의 합계 출력

print(np.mean([1,2,3]));    #배열 요소의 평균

print(np.std([1,2,3]));     #배열 요소의 표준편차
```

Pandas 함수

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#당뇨병 데이터 불러오기
df=pd.read_csv('dataset/diabetes.csv');
print(df.head());

print(df.tail());

print(df.info());

print(df.describe());
```

```python
#각 변수의 분포 확인
df.hist(figsize=(10,8))
plt.show()
```

```python
#변수 간 상관관계 시각화
sns.heatmap(df.corr(),annot=True)
plt.show()
```

```python
#결측치 확인(null 확인)
df.isnull().sum()
```

```python
#결측치 평균값으로 채우기
df.fillna(df.mean(),inplace=True)       #inplace=True는 변경된 설정으로 원본 데이터에 덮어쓴다는 의미
df
```

Matpltlib 함수

```python
import matplotlib.pyplot as plt
import numpy as np

#데이터 생성
x=np.linspace(0,10,100);
y=np.sin(x);

#선 그래프
plt.plot(x,y);
plt.xlabel('x');
plt.ylabel('sin(x)');
plt.title('SineWave');
plt.show()
```

```python
#산점도
plt.scatter(x,y)
plt.show()
```

```python
#막대 그래프
data=[3,8,1,10];
plt.bar(["A","B","C","D"],data);
plt.show();
```

```python
#히스토그램
np.random.seed(19680801);
data=np.random.randn(1000);
plt.hist(data,bins=30);
plt.show();
```

Seaborn 함수

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

#팁 데이터 불러오기
tips=sns.load_dataset("tips");

#팁 금액 분포 시각화
sns.distplot(tips['tip'],kde=True);
plt.show();
```

```python
#성별에 따른 팁 금액 비교(박스플롯)
sns.boxplot(x="sex",y="tip",data=tips);
plt.show();
```

```python
#팁 금액과 총 지출액 간의 관계(산점도)
sns.scatterplot(x="total_bill",y="tip",data=tips);
plt.show();
```
